```python
import numpy as np
import matplotlib.pyplot as plt
x = np.logspace(0,15,100,dtype=np.float128) 
```

```python
y_1 = np.ones(100)
y_n = x
y_square = x*x
y_squareroot = np.sqrt(x)
y_nlogn = x*np.log(x)
y_cube = x**3
y_2n = 2**x
```

```python
fig = plt.figure()
p = fig.add_subplot(1,1,1)
p.set_yscale('log')
p.set_xscale('log')
p.set_xlim(1,10**15)
p.set_ylim(1,10**19)
p.plot(x,y_1,label='1',color = 'black')
p.plot(x,y_n,label='n',color = 'green')
p.plot(x,y_square,label='y_square',color = 'cyan')
p.plot(x,y_squareroot,label='y_squareroot',color = 'blue')
p.plot(x,y_nlogn,label='y_nlogn',color = 'purple')
p.plot(x,y_cube,label='y_cube',color = 'magenta')
p.plot(x,y_2n,label='y_2n',color = 'silver')
p.legend(['1','n','square','sqrroot','nlogn','cube','2n'])
p.plot()
```
